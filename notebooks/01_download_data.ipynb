```python
import yfinance as yf
import numpy as np
import pandas as pd

nvda = yf.download(
    "NVDA",
    start="2000-01-01",
    end="2026-01-01",
    progress=False
)

returns = np.log(nvda["Close"]).diff().dropna()

print("Price observations:", len(nvda))
print("Return observations:", len(returns))

nvda.to_csv("../data/raw/nvda_prices_2000_2026.csv")
returns.to_csv("../data/processed/nvda_log_returns_2000_2026.csv")
```
